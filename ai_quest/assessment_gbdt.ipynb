```python
DEBUG = False
if DEBUG:
    train_df = train_df.sample(frac = 0.1).reset_index(drop = True)
    y_df = y_df.sample(frac=0.1).reset_index(drop = True)
    print(train_df.shape, y_df.shape)
```

```python
train_X, test_X, train_y, test_y = train_test_split(train_df, y_df, test_size=0.2)
print(len(train_X),len(train_y),len(test_X),len(test_y))
```

```python
# モデルの学習
GBDT = GradientBoostingRegressor()
GBDT.fit(train_X, train_y)

# 回帰　
pred_GBDT = GBDT.predict(test_X)

# 評価
# 決定係数(R2)
r2_GBDT = r2_score(test_y, pred_GBDT)

# 平均絶対誤差(MAE)
mae_GBDT = mean_absolute_error(test_y, pred_GBDT)

print("R2 : %.3f" % r2_GBDT)
print("MAE : %.3f" % mae_GBDT)

# 変数重要度
print("feature_importances = ", GBDT.feature_importances_)
```
